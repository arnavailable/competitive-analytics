```python
from itertools import product
from matplotlib import pyplot as plt
import numpy as np
N = 10

# Generate the data
def f(i, j, k):
    def mod(x):
        return x if x > 0 else -x
    share = 0
    for l in range(1, N+1):
        if mod(j-l) == mod(i-l) and mod(i-l) == mod(k-l):
            share += 10/3
        elif mod(j-l) == mod(i-l) and mod(i-l) < mod(k-l):
            share += 10/2
        elif mod(k-l) == mod(i-l) and mod(i-l) < mod(j-l):
            share += 10/2
        elif mod(i-l) < mod(j-l) and mod(i-l) < mod(k-l):
            share += 10
        else:
            share += 0
    return share

volume = np.zeros((N, N, N))
for i in range(N):
    for j in range(N):
        for k in range(N):
            volume[i, j, k] = f(i+1, j+1, k+1)

# Plot the 3-D data
fig = plt.figure(figsize=(15,15))
ax = fig.add_subplot(projection="3d")
space = np.array([*product(range(1, N+1), range(1, N+1), range(1, N+1))]) # all possible triplets of numbers from 0 to N-1 # generate random data
ax.scatter(space[:,0], space[:,1], space[:,2], c=(100-volume)/100, s=volume**2, alpha=1) # plot the points
ax.set_title('3 Candidate Election')
ax.set_xticks(range(1, N+1))
ax.set_yticks(range(1, N+1))
ax.set_zticks(range(1, N+1))
ax.set_xlabel('Voter 1', labelpad=5)
ax.set_ylabel('Voter 2', labelpad=5)
ax.set_zlabel('Voter 3', labelpad=5)
```
